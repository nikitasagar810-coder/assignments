# BASIC STATISTICS

# 1. Descriptive Analytics for Numerical Columns
 ## Objective:

To compute and analyze basic statistical measures such as mean, median, mode, and standard deviation for the numerical columns in the dataset.

## Steps:

Load the dataset using Python (Pandas).

Identify the numerical columns (e.g., Sales Amount, Discount %, Quantity, Profit, etc.).

Calculate the following for each numerical variable:

Mean

Median

Mode

Standard Deviation

Provide interpretations based on these values.

# Load the dataset using Python (Pandas).

```python
import pandas as pd
df=pd.read_csv("sales_discounts.csv")
```

```python
df
```

```python
num_cols = df.select_dtypes(include=['int64','float64'])
```

```python
num_cols
```

```python
mean_values=num_cols.mean()
mean_values
```

```python
median_values=num_cols.median()
median_values
```

```python
mode_values=num_cols.mode()
mode_values
```

```python
std_values=num_cols.std()
std_values
```

##Interpretation Example

Sales Amount may have a high standard deviation, indicating large variability in sales.

Discount % may show low mean values, implying discounts are rarely high.

If median < mean, the column is positively skewed, meaning a few high values boost the mean.

Mode helps show the most common discount or quantity purchased.

## 2. Data Visualization
A. Histograms (Numerical Variables)
  ### Objective:

To observe the distribution, skewness, and possible outliers.

```python
import matplotlib.pyplot as plt
```

```python
num_cols.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

## Inference from Histograms (Example)

Sales values may be right-skewed, meaning a few transactions have very high sales.

Quantity purchased may cluster around small numbers (1 to 3).

Discount percentages may be mostly zero, with a few high discount cases.

## B. Box Plots
Objective:

To detect outliers and understand the interquartile range (IQR).

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(data=num_cols)
plt.show()
```

## Findings Example

Sales and Profit typically show several outliers, representing unusually high transactions.

Discount % may show no extreme outliers, indicating consistent discounting.

IQR helps identify whether the data is tightly packed or widely spread.

## C. Bar Chart Analysis for Categorical Columns
Objective:

To visualize the frequency of different categories.

Steps:

Identify categorical columns (e.g., Region, Category, Product Type).

Plot bar charts for each categorical variable.

```python
cat_cols = df.select_dtypes(include=['object'])
```

```python
cat_cols
```

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

## Insights Example

The Category column may show that some product categories (like Electronics or Furniture) dominate sales.

Region frequency distribution may show more sales in urban markets compared to rural ones.

Helps identify which customer segments are most active.

# Conclusion

From the descriptive analytics and visualization process, we observe the following:

Numerical columns display clear patterns such as skewness, high variability, and the presence of outliers.

Histograms highlight that sales and discounts are not uniformly distributed—sales often have a long tail.

Box plots reveal significant outliers in high sales or profit transactions.

Categorical data analysis shows which product categories or regions contribute most to sales activity.

These findings help in decision-making, sales forecasting, and discount optimization for future analysis.
