### 1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list
l1 = [34, 46, 88, 90, 94]
print("Original list:", l1)
```

```python
#append
l1.append (98) # Adds an element to the end of the list
print ("After append",l1)
```

```python
#extend
l1.extend([87,97]) #Adds multiple elements from an iterable to the end of the list.
print(  "After extend",l1)
```

```python
#insert
l1.insert(2,95) #insert an element at a specified index.
print("After insert",l1)
```

```python
#remove
l1.remove(94) #remove the first occurrence of an element
print("After remove",l1)
```

```python
# pop() - Removes element by index (default last element)
popped_value = l1.pop()
print("After pop():", l1)
print("Popped Value:", popped_value)
```

```python
# index() - Returns index of a given element
index_of_88 = l1.index(88)
print("Index of 88:", index_of_88)
```

```python
# count() - Counts how many times a value appears
count_46 = l1.count(46)
print("Count of 20:", count_46)
```

```python
# sort() - Sorts the list
l1.sort()
print("After sort():", l1)
```

```python
#revers() - reverses the list in place 
l1.reverse()
print("After revers",l1)
```

```python
# clear() - Removes all elements
l1.clear()
print("After clear():", l1)

```

### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows



```python
# Creating the DataFrame with 10 rows
import pandas as pd
import numpy as np

data = {
    "Name": ["Nikita","Nitya","Abhi","Keerti","Ankita","Laxmi","Shweta","Nikhil","Avi","sagar"],
    "Age": [20,6,22,21,19,18,17,30,28,18],
    "City": ["chitta","mubai","Abulga","Basavakalyan","Bengular","pune","Musirabad","Nilanga","Gulbarga","Pune"],
    "Salary":[50000,40000,30000,np.nan,30000,80000,10000,5000,np.nan,7000]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Use .info()
df.info()
```

```python
#Use.describe()
df.describe()
```

```python
#Select 'Name' and 'City' columns
df[['Name','City']]
```

```python
#Drop 'City' column
df.drop('City', axis=1)
```

```python
#Fill missing values in 'Salary' column with the mean

df['Salary'].mean()
```

```python
df['Salary'].mean()
df['Salary'].fillna(df['Salary'].mean())
```

```python
 #Remove duplicate rows
df = df.drop_duplicates()
```
