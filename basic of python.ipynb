## Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
# Step 1: Input from the User

num= int(input("enter a number:"))
```

the program asks the user to enter a number

 input() taskes the values as a string.

 int() convers it into an interger and stores it in the variable num

```python
# prime numbers are greater than 1
if num > 1:
    # check for factors
    for i in range(2, int(num**0.5) + 1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

How it works
1. Taking user input
2. Checking if it is greater than 1
3. Testing divisibility from 2 to num
4. If divisible -> Not prime
5. If no divisor -> prime

## Exercise 2: Product of Random Numbers
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message


```python
# step 1: Importing the random module
import random
```

This allows us to generate random numbers.

```python
# Step 2: Generate two random number
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
```

randint(1,10) generates a number between 1 and 10.

```python
# Step 2: Show the numbers to the user
print("Number 1:", num1)
print("Number 2:", num2)
```

```python
# Step 3: Ask user for the product
user_answer = int(input("Enter the product of these two numbers: "))
```

input() reads user input.

int() converts it to an integer.

```python
# Step 4: Calculate the correct product
correct_product = num1 * num2

```

```python

# Step 5: Check the answer
if user_answer == correct_product:
    print("Correct! Well done!")
else:
    print("Incorrect. The correct answer is:", correct_product)
```

```python
  # Squares of  Even Numbers
# We are choosing EVEN numbers from 100 to 200

for num in range(100, 201):
    if num % 2 == 0:  # Check if the number is even
        print(f"{num} squared is {num ** 2}")
```

```python
 #Squares of Odd Numbers
# We are choosing ODD numbers from 100 to 200

for num in range(100, 201):
    if num % 2 != 0:  # Check if the number is odd
        print(f"{num} squared is {num ** 2}")
```

 ## How it works

1. range(100, 201) → gives numbers from 100 to 200

2. num % 2 == 0 → checks if number is even

3.  num % 2 != 0 → checks if number is odd

4. num ** 2 → calculates square

5. print() → displays the result

## Exercise 4: ✅ Problem Statement (Question):
Hospital Billing System A hospital charges its patients based on the following:
• Room charges per day: ₹2,000
• Doctor consultation fee (flat): ₹1,500 
• Lab test charges: Based on the number of tests taken. Each test costs ₹300
• Medicine charges: Total cost of medicines provided
• Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
• Stayed for 4 days
• Had 3 lab test


```python
# Given data
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300
medicine_charges = 0  # You can change this later if needed
```

```python
# Patient details
days_stayed = 4
number_of_tests = 3
```

```python
# Calculating room charges
room_total = room_charge_per_day * days_stayed
print("room_total",room_total)
```

```python
# Calculating lab test charges
lab_total = lab_test_cost * number_of_tests
print("lab_total",lab_total)
```

```python
# Total before discount

total_before_discount = room_total + doctor_fee + lab_total + medicine_charges
print("total_before_discount",total_before_discount)
```

```python
# Applying discount if total exceeds 10,000
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
  discount = 0
```

```python
# Final bill after discount
final_bill = total_before_discount - discount
```

```python
# Printing the final bill
print("Total Bill Amount: ₹", final_bill)
```
